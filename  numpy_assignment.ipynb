### 1(A) Calculate the sum, mean, median, and standard deviation of the elements in the array.  

```python
list1 =list(range(1,21))
arr = np.array(list1)

arr_sum = int(sum(arr))
arr_mean = float(np.mean(arr))
arr_median = float(np.median(arr))
arr_std = float(np.std(arr))
print(arr)
print(f"Sum = {arr_sum}")
print(f"Mean = {arr_mean}")
print(f"Median = {arr_median}")
print(f"standard deviation = {arr_std}")
```

## 2. Create a 2D NumPy array of shape 4 X 4 with numbers ranging from 1 to 16. ##2(A) Print the array.
### 2(A) Print the array. 

```python
arr_2d = np.arange(1,17).reshape(4,4)
print(f" 1 to 16 4X4 arr ={arr_2d}")
```

### 2(B)Find the transpose of the array.  

```python
print(f"transpose of the array:{arr_2d.T}")
```

### 2(C) Calculate the row-wise and column-wise sums of the array.  

```python
print(f"row-wise Sum={np.sum(arr_2d,axis=1)}")
print(f"column-wise Sum={np.sum(arr_2d,axis=0)}")
```

## 3. Create two 3 X 3 arrays filled with random integers between 1 and 20.  
### 3(A) Perform element-wise addition, subtraction, and multiplication.  

```python
arr3 =np.arange(1,10).reshape(3,3)
arr4 =np.arange(11,20).reshape(3,3)
print(arr3)
print(arr4)

print(f"addition={arr3+arr4}")
print(f"subtraction={arr3-arr4}")
print(f"multiplication={arr3*arr4}")
```

### 3(B) Compute the dot product of the two arrays.  

```python
print(f"dot product of the two arrays={np.dot(arr3, arr4)}")
```
