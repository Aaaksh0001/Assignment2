```python
import pandas as pd
```

## 1. Create a DataFrame with the following data:  


```python

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
```

```python
df
```

###  1(a). Print the first five rows of the DataFrame. 

```python
df.head()
```

###  1(b). Get the summary statistics of the 'Age' and 'Salary' columns.  

```python
df[['Age', 'Salary']].describe()
```

### 1(c). Calculate the average salary of employees in the 'HR' department. 

```python
float(df[df['Department'] == 'HR']['Salary'].mean())
```

## 2. Add a new column, 'Bonus', which is 10% of the salary. 

```python
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

## 3. Filter the DataFrame to show employees aged between 25 and 30.

```python
df[(df['Age'] >= 25) & (df['Age'] <= 30)]
```

## 4. Group the data by 'Department' and calculate the average salary for each department.  

```python
df.groupby('Department')['Salary'].mean()
```

## 5. Sort the DataFrame by 'Salary' in ascending order and save the result to a new CSV file. 

```python
sort_df = df.sort_values(by ='Salary' , ascending = True)
```

```python
sort_df
```

```python
sort_df.to_csv("newcsv.csv",index = False)
print("create new CSV file successfully")
```
