## 1. Create a simple line plot for the following data: 
###  a. Plot the data.  
###  b. Customize the plot by adding a title, axis labels, and a grid.  


```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x,y)
plt.xlabel("X_axis")
plt.ylabel("y_axis")
plt.title("Line Plot")
plt.grid()
plt.show()
```

## 2. Create a bar graph to represent the marks scored by students in a subject: 
### a. Plot the data as a bar graph.  
###  b. Customize the colors and add a title.  

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
color=["r","b","g","y"]
plt.bar(students,marks,color=color)
plt.xlabel("Studens")
plt.ylabel("Marks")
plt.title("Bar Graph Student Marks Scord ")
```

## 3. Create a pie chart to represent the percentage distribution of a companys revenue from different regions:  
###  a. Create a pie chart with the region names as labels.  
###  b. Highlight the region with the highest revenue.  


```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',  startangle=90)
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
max_index = revenue.index(max(revenue))

explode = [0.1 if i == max_index else 0 for i in range(len(revenue))]
plt.pie(revenue,explode=explode,labels=regions,autopct='%1.1f%%',  startangle=90)
```

## 4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).  

```python
import numpy as np
```

```python
rend = np.random.randint(1,100,1000)
```

```python
plt.hist(rend,bins=20)
plt.xlabel("X_axis")
plt.ylabel("y_axis")
plt.title("Histogram Plot")
plt.show()
```
